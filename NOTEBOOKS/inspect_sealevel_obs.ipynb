```python
import xarray as xr
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os
```

```python
mooring = 'W310'
# mooring = 'S245'
ds = xr.open_dataset(f'~/data/ProjectsActive/SWOT_NAust/DATA/mooring_sealevel_{mooring}_v2.nc')
ds
```

```python
def remove_labels(ax):
    ax.set_title('')
    ax.set_xlabel('')
    ax.set_xticklabels([])



def plot_fig(xlims, ylims,c):
    fig = plt.figure(figsize=(8,5))
    
    ax0=plt.subplot(411)
    (ds['eta']-ds['eta_tide_fes']).plot(c=c, lw=0.5)
    remove_labels(ax0)
    ax0.set_xlim(xlims)
    ax0.set_ylim(ylims)
    plt.grid(True, ls=':')
    ax0.set_ylabel("$η'$ [m]")
    
    ax1=plt.subplot(412)
    (ds['eta_b']-ds['eta_tide_fes']).plot(c=c, lw=0.5)
    remove_labels(ax1)
    ax1.set_xlim(xlims)
    ax1.set_ylim(ylims)
    plt.grid(True, ls=':')
    ax1.set_ylabel("$η_b'$ [m]")
    
    
    ax2=plt.subplot(413)
    (ds['eta_steric']).plot(c=c, lw=0.5)
    remove_labels(ax2)
    ax2.set_xlim(xlims)
    ax2.set_ylim(ylims)
    plt.grid(True, ls=':')
    ax2.set_ylabel("$η_s$ [m]")
    
    
    ax3=plt.subplot(414)
    (ds['eta_a']).plot(c=c, lw=0.5)
    ax3.set_xlim(xlims)
    ax3.set_ylim(ylims)
    plt.grid(True, ls=':')
    ax3.set_ylabel("$η_a$ [m]")
    ax3.set_title('')
    
    plt.subplots_adjust(top = 1, bottom = 0, right = 1, left = 0, 
            hspace = 0, wspace = 0)
    plt.tight_layout()
```

```python
xlims = (ds.time[0], ds.time[-1])
ylims = (-0.35, 0.35)
c = 'b'
plot_fig(xlims, ylims, c)
outfile=f'../FIGURES/sealevel_{mooring}.png'
plt.savefig(outfile)
outfile
```
